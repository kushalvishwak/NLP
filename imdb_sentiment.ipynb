```python
df = pd.read_csv("imdb_sentiment.csv")
df.head()
```

```python
google_model = 'D:\\unstructure data IMDB\\GoogleNews-vectors-negative300.bin.gz'
embeddings = gensim.models.KeyedVectors.load_word2vec_format(google_model,binary=True)
```

####Clean the data set:
    1. retain only alphabets and numbers
    2. removestopwords , puntations and worlds like -
    would ,could , told ,subject
    3. retain words with langth above 2

```python
embeddings['google'].shape
```

```python
stop_nltk = stopwords.words("english")
```

```python
stop_updated = stop_nltk + ["...","..","!!"] +["would" ,"could" , "told" ,"subject"]
```

```python
print(stop_nltk)
```

```python
from string import punctuation
from nltk.tokenize import word_tokenize
```

```python
def  clean_txt(sent):
    tokens = word_tokenize(sent.lower())
    stemmed = [term for term in tokens 
               if term not in stop_updated and
               term not in list(punctuation) and len(term) > 2] 
    res = " ".join(stemmed)
    return res
```

```python
df["clean_review"]=df.review.apply(clean_txt)
```

```python
df.head()
```

#### create document vector using word embeddings

```python
docs_vectors = pd.DataFrame()
for doc in df.clean_review:
    temp = pd.DataFrame()
    #print(temp)
    words = doc.split(' ')
    #print(words)
    for word in words:
        try:
            words2vec = embeddings[word]   ## gives vector of the length
            temp = temp.append(pd.Series(words2vec),ignore_index=True)
        except:
            pass
    doc_vector = temp.mean()
    print(doc_vector)
    docs_vectors = docs_vectors.append(doc_vector,ignore_index=True)
```

```python
docs_vectors=docs_vectors.fillna(method='ffill')
```
